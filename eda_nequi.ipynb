# Limpieza y transformación de datos

Se borran las filas con nulos y se hacen las transformaciones necesarias para que los datos sean más fáciles de manejar. Además, se rellena la columna de "keywords" usando la columna "description".

```python
# Eliminar filas con valores nulos en la columna total_views, total_videos y description
df.dropna(subset=['total_views', 'total_videos', 'description'], inplace=True)
```

```python
# Mostrar número de filas y columnas después de eliminar valores nulos
df.shape
```

```python
# Formateo de las columnas

# Convertir a datetime
df['join_date'] = pd.to_datetime(df['join_date'])

# Convertir a string
df['channel_id'] = df['channel_id'].astype(str)
df['channel_link'] = df['channel_link'].astype(str)

# Convertir a int
df['subscriber_count'] = df['subscriber_count'].astype(int)
df['total_videos'] = df['total_videos'].astype(int)
df['total_views'] = df['total_views'].astype(int)
```

```python
# Mostrar las columnas y sus tipos de datos después del formateo
df.info()
```

```python
# Mostrar las keywords más repetidas en los top 10 canales
def count_keywords(keywords_list):
    keyword_count = {}
    for keywords in keywords_list:
        for keyword in keywords.split(','):
            keyword = keyword.strip().lower()  # Convertir a minúsculas y eliminar espacios alrededor
            keyword_count[keyword] = keyword_count.get(keyword, 0) + 1
    return keyword_count

# Calcular el top 10 de canales por suscriptores
top_channels = df.nlargest(10, 'subscriber_count')

# Calcular las palabras clave en los top 10 canales
all_keywords = top_channels['keywords'].tolist()
keyword_count = count_keywords(all_keywords)

# Ordenar las palabras clave por frecuencia y obtener las top 10
top_keywords = sorted(keyword_count.items(), key=lambda x: x[1], reverse=True)[:10]

print("Top 10 palabras clave:")
for keyword, count in top_keywords:
    print(f"{keyword}: {count}")
```

```python
# Extraer keywords de la columna 'description'
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Cargar el dataset
df = pd.read_csv('c:/Users/Andrea bb/Downloads/Spotify/youtube_channels_1M_clean.csv')

# Eliminar filas con valores nulos en la columna 'description'
df_filtered = df.dropna(subset=['description'])

# Definir función para tokenizar y limpiar texto
def process_text(text):
    # Tokenizar el texto en palabras
    tokens = word_tokenize(text.lower())
    # Eliminar signos de puntuación y números
    tokens = [word for word in tokens if word.isalpha()]
    # Eliminar stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lematizar las palabras
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

# Aplicar la función a la columna 'description'
df_filtered['tokens'] = df_filtered['description'].apply(process_text)

# Extraer palabras clave de los tokens
df_filtered['keywords'] = df_filtered['tokens'].apply(lambda x: ' '.join(x[:5]))

# Mostrar el DataFrame con los datos imputados
print(df_filtered[['description', 'keywords']].head())
```

```python
# Gráfico de correlación
import matplotlib.pyplot as plt
import seaborn as sns
# Gráfico de correlación
# Seleccionar solo las columnas numéricas
numeric_columns = df.select_dtypes(include=['float64', 'int32', 'int64']).columns
numeric_df = df[numeric_columns]

# Calcular la matriz de correlación
correlation_matrix = numeric_df.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre variables')
plt.show()
```
